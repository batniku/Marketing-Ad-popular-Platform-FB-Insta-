```python
slices=data.groupby('campaign_name')['clicks'].sum()
slices
```

```python
cols=['c','m','r','g']
campaign_name=slices.index

plt.pie(slices,
       labels=campaign_name,
       colors=cols,
       startangle=90,
       shadow=True,
       explode=(0,0,0.1,0),
       autopct='%1.1f%%')

plt.title('Highest campaign_no')
plt.legend()
```
